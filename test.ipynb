```python
import numpy as np

lr = 0.1

N = 100

X = np.array([7.25, 2.64, 0.32, 5.5, 16.45, 2.888, 0.01, 2.1, 0.7, 0.4, 0.2, 4.4, 2.7, 3.3, 2.3]).reshape(5, 3)
y = np.array([1.0, 0.0, 0.0, 1.0, 1.0])

def sigmoid(x : np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-x))

w = np.array([0.0, 0.0, 0.0])
b = 0.0

for _ in range(N):

    y_hat = sigmoid(X @ w + b)
    dz = y_hat - y

    dw = (X.T @ dz) / X.shape[0]
    db = dz.sum() / X.shape[0]

    w -= lr * dw
    b -= lr * db
```
